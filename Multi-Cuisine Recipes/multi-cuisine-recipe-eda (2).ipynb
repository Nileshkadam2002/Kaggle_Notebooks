# Which cuisines require more ingredients (higher cooking effort)?

```python

df['ingredient_count']=df['ingredients'].apply(lambda x:len(x.split(',')))
df.groupby('area')['ingredient_count'].mean().sort_values(ascending=False)
```

# Which cuisines are easier for beginners?

```python
df['num_steps']=df['steps'].apply(lambda x:len(x.split('.')))
df.groupby('area')['num_steps'].mean()
```

# What type of dishes are popular within each cuisine?

```python
pd.crosstab(df['area'],df['category'])
```

```python
supply=df['area'].value_counts()
print(supply.tail(10))
```

# Identified 65% recipe concentration in top 5 cuisines

```python
top5=df['area'].value_counts().head(5)
top5_share=(top5.sum()/len(df)*100)
print(f"Top 5 Cuisine contribute {top5_share:2f}% of total recipes" )
```

# Detected underrepresented cuisines for catalog expansion

```python
cuisine_counts=df['area'].value_counts()
underrepresented=cuisine_counts[cuisine_counts < cuisine_counts.mean()]
print(underrepresented)
```

# Heatmap (Cuisine vs Category)

```python
import seaborn as sns
import pandas as pd
pivot=pd.crosstab(df['area'],df['category'])

plt.figure(figsize=(12,8))
sns.heatmap(pivot,cmap='YlGnBu')
plt.title('Cuisine vs Category Diversirty ')
plt.show()
```

```python
df.columns
```

```python
premium = ['cheese','salmon','wine','avocado']

df['premium_recipe'] = df['ingredients'].str.lower().str.contains('|'.join(premium))
```

```python
df['premium_recipe'].value_counts()
```

```python
non_veg_words = ['chicken','fish','mutton','beef','egg']

df['is_vegetarian'] = ~df['ingredients'].str.lower().str.contains('|'.join(non_veg_words))
```

```python
df['is_vegetarian'].value_counts()
```

```python
veg_counts=df['is_vegetarian'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(veg_counts,labels=['Vegetarian','Non- Vegetarian'],autopct='%1.1f%%',colors=['#2ecc71', '#e74c3c'],startangle=140)
plt.title('Proportion of Vegetarian  vs Non Vegetarian Recipes')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='is_vegetarian', palette=['#e74c3c', '#2ecc71'])
plt.title('Count of Veg vs Non-Veg Recipes')
plt.xlabel('Recipe Type')
plt.ylabel('Number of Recipes')
plt.xticks([0, 1], ['Non-Vegetarian', 'Vegetarian'])
plt.show()
```

```python
diet_by_cuisine = pd.crosstab(df['area'], df['is_vegetarian'])
diet_by_cuisine.columns = ['Non-Veg', 'Veg']
diet_by_cuisine.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#e74c3c', '#2ecc71'])
plt.title('Vegetarian vs Non-Vegetarian Recipes by Cuisine')
plt.ylabel('Number of Recipes')
plt.show()
```
